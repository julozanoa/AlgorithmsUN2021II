#**HeapSort**

In computer science, heapsort is a comparison-based sorting algorithm. Heapsort can be thought of as an improved selection sort: like selection sort, heapsort divides its input into a sorted and an unsorted region, and it iteratively shrinks the unsorted region by extracting the largest element from it and inserting it into the sorted region. Unlike selection sort, heapsort does not waste time with a linear-time scan of the unsorted region; rather, heap sort maintains the unsorted region in a heap data structure to more quickly find the largest element in each step

#Experimental analysis of the Heap Sort execution time

```python
# HeapSort Code
def max_heapify(a, n, i):
  max = i
  left = 2 * i + 1
  right = 2 * i + 2

  if left < n and a[i] < a[left]:
    max = left
  
  if right < n and a[max] < a[right]:
    max = right
  
  if max != i:
    a[i], a[max] = a[max], a[i]
    max_heapify(a, n, max)

def build_max_heap(a, n):
  begin = (n // 2) -1
  end = -1
  for i in range(begin, end, -1):
    max_heapify(a, n, i)

  
def heapSort(a):
    n = len(a)
    begin = (n // 2) - 1
    end = -1
    build_max_heap(a, n)

    for i in range(n-1, 0, -1):
      a[i], a[0] = a[0], a[i]
      max_heapify(a, i, 0)
```

```python
'''Test 1'''
import numpy as np
import matplotlib.pyplot as plt
import time 
tam = np.arange(1,200,1)
exit_heapsort = []
for i in tam:
    l = list(range(i))
    x = time.time() 
    heapSort(l)
    y = time.time() 
    sum = y - x
    exit_heapsort.append(sum) 
plt.plot(tam, exit_heapsort, 'bx', label = 'HeapSort', color='green')
plt.xlabel('Array size')
plt.ylabel('Seconds')
plt.legend(loc = 'upper left')
```

```python
# HeapSort Code Steps
steps = 0

def max_heapify_steps(a, n, i):
  global steps
  max = i
  left = 2 * i + 1
  right = 2 * i + 2

  if left < n and a[i] < a[left]:
    max = left
    steps += 1
  
  if right < n and a[max] < a[right]:
    max = right
    steps += 1
  
  if max != i:
    a[i], a[max] = a[max], a[i]
    max_heapify_steps(a, n, max)
    steps += 2

  steps += 6

def build_max_heap_steps(a, n):
  global steps
  begin = (n // 2) -1
  end = -1
  for i in range(begin, end, -1):
    max_heapify_steps(a, n, i)
    steps += 1
  steps += 3
  
def heapSort_steps(a):
    global steps
    steps = 0
    n = len(a)
    begin = (n // 2) - 1
    end = -1
    build_max_heap_steps(a, n)

    for i in range(n-1, 0, -1):
      a[i], a[0] = a[0], a[i]
      max_heapify_steps(a, i, 0)
      steps += 2

    steps += 5
    return steps
```
