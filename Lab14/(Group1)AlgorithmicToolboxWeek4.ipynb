```python
def bs(arr, ini, fnl, x):
 
    # Check base case
    if fnl >= ini:
 
        mid = ini + (fnl - ini) // 2
        
        if arr[mid] == x:
            return mid
 
        elif arr[mid] > x:
            return bs(arr, ini, mid-1, x)
 
        else:
            return bs(arr, mid + 1, fnl, x)
 
    else:
        return -1
 
 
# lectura y ejecuci√≥n
cantidad = int(input())
arr = input()
cantidad2 = int(input())
arr2 = input()

lista_datos = list(map(lambda x: int(x), arr.split() ))
lista_datos_a_encontrar = list(map(lambda x: int(x), arr2.split() ))

resultado = list()
for i in lista_datos_a_encontrar:
    resultado.append(bs(lista_datos,0,(cantidad-1),i))

for i in resultado:
    print (i)
```

```python
def bs(arr,x):
    
    low = 0
    high = len(arr)-1
    
    while low <= high:
        mid = (low + high) // 2
        
        if arr[mid] < x:
            low = mid + 1
        elif arr[mid] > x:
            high = mid - 1
        else:
            if mid - 1 < 0:
                return mid
            if arr[mid - 1] != x:
                return mid
            high = mid - 1
    
    return -1
 
# lectura y ejecuci√≥n
cantidad = int(input())
arr = input()
cantidad2 = int(input())
arr2 = input()

lista_datos = list(map(lambda x: int(x), arr.split() ))
lista_datos_a_encontrar = list(map(lambda x: int(x), arr2.split() ))

resultado = list()
for i in lista_datos_a_encontrar:
    resultado.append(bs(lista_datos,i))


for i in resultado:
    print (i, end= " ")
```

```python
n = int(input())
seq = [int(i) for i in input().split()]


def divide_func(seq, l, r):
    if l+1==r:
        return seq[l]
    elif l+2==r:
        return seq[l]
    m = (l+r)//2
    left = divide_func(seq, l, m)
    right = divide_func(seq, m, r)

    c1, c2 = 0, 0
    for i in seq[l:r]:
        if i == left:
            c1+=1
        elif i == right:
            c2+=1
    if c1>(r-l)//2 and left != -1:
        return left
    elif c2>(r-l)//2 and right != -1:
        return right
    else: 
        return -1

print(int(divide_func(seq, 0, n) != -1))
```
