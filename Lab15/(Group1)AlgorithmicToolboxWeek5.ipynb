```python
import numpy


def edit_distance(A, B):
    m, n = len(A), len(B)
    d = numpy.zeros((m + 1, n + 1), dtype=int)
    
    for i in range(n + 1):
        d[0, i] = i
        
    for i in range(m + 1):
        d[i, 0] = i
        
    for j in range(1, n + 1):
        for i in range(1, m + 1):
            insertion = d[i, j - 1] + 1
            deletion = d[i - 1, j] + 1
            match = d[i - 1, j - 1]
            mismatch = d[i - 1, j - 1] + 1
            
            if A[i - 1] == B[j - 1]:
                d[i, j] = min(insertion, deletion, match)
            else:
                d[i, j] = min(insertion, deletion, mismatch)
                
    return d[m, n]


print(edit_distance(input(), input()))
```

```python
import math


def money_change_a(m, d, min_c):
    for i in range(1, m + 1):
        for j in d:
            if i >= j:
                coins = min_c[i - j] + 1
                if coins < min_c[i]:
                    min_c[i] = coins
    
    return min_c


money = int(input())
denominations = [1, 3, 4]
min_coins = [0] + [math.inf] * money

print(money_change_a(money, denominations, min_coins)[money])
```
